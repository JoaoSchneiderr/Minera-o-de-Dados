### 2. Gravidade dos acidentes por mês em Nova Iguaçu


```python
nova_iguacu_df['data_inversa'] = pd.to_datetime(nova_iguacu_df['data_inversa'])
nova_iguacu_df['mês'] = nova_iguacu_df['data_inversa'].dt.month

gravidade_mensal = nova_iguacu_df.groupby('mês')[['mortos', 'feridos_graves', 'feridos_leves', 'ilesos']].sum()

print("\nAnálise da Gravidade dos Acidentes por Mês em Nova Iguaçu (Média Mensal):")
avg_mortos_mes = gravidade_mensal['mortos'].mean()
avg_feridos_graves_mes = gravidade_mensal['feridos_graves'].mean()
avg_feridos_leves_mes = gravidade_mensal['feridos_leves'].mean()
print(f"Em média, mensalmente, ocorreram {avg_mortos_mes:.1f} mortes, {avg_feridos_graves_mes:.1f} feridos graves e {avg_feridos_leves_mes:.1f} feridos leves.")

total_mortos_geral = gravidade_mensal['mortos'].sum()
total_feridos_graves_geral = gravidade_mensal['feridos_graves'].sum()
total_feridos_leves_geral = gravidade_mensal['feridos_leves'].sum()
total_ilesos_geral = gravidade_mensal['ilesos'].sum()

print("\nTotais Gerais e Proporções (Ano de 2025) ")
total_mortos = total_mortos_geral
total_feridos = total_feridos_graves_geral + total_feridos_leves_geral
print(f"Total Geral de Mortos: {total_mortos:.0f}")
print(f"Total Geral de Feridos (Graves + Leves): {total_feridos:.0f}")
```

```python
classificacao_acidente_contagem = nova_iguacu_df['classificacao_acidente'].value_counts()

acidentes_com_vitimas_fatal = classificacao_acidente_contagem.get('Com Vítimas Fatais', 0)
acidentes_com_vitimas_feridas = classificacao_acidente_contagem.get('Com Vítimas Feridas', 0)
acidentes_sem_vitimas = classificacao_acidente_contagem.get('Sem Vítimas', 0)

total_registros_acidentes = len(nova_iguacu_df)

prop_com_vitimas = ((acidentes_com_vitimas_fatal + acidentes_com_vitimas_feridas) / total_registros_acidentes) * 100
prop_sem_vitimas = (acidentes_sem_vitimas / total_registros_acidentes) * 100

print(f"Proporção de registros de acidentes com vítimas (fatais ou feridas): {prop_com_vitimas:.2f}%")
print(f"Proporção de registros de acidentes sem vítimas: {prop_sem_vitimas:.2f}%")
```

```python
taxa_acidentes_morte = (acidentes_com_vitimas_fatal / total_registros_acidentes) * 100
print(f"Taxa de registros de acidentes com vítimas fatais: {taxa_acidentes_morte:.2f}%")
```

```python
taxa_acidentes_morte = (acidentes_com_vitimas_fatal / total_registros_acidentes) * 100
print(f"Taxa de registros de acidentes com vítimas fatais: {taxa_acidentes_morte:.2f}%")


total_unique_accidents = nova_iguacu_df['id'].nunique()
total_vitimas = total_mortos_geral + total_feridos_graves_geral + total_feridos_leves_geral
media_vitimas_por_acidente = total_vitimas / total_unique_accidents
print(f"Média de vítimas (mortos + feridos graves + feridos leves) por acidente único: {media_vitimas_por_acidente:.2f}")
```

### 9. Análise do Acidente Mais Grave







```python
acidentes_analise = (
    nova_iguacu_df.groupby('id')
    .agg(
        total_mortos=('mortos', 'sum'),
        total_feridos_graves=('feridos_graves', 'sum'),
        total_feridos_leves=('feridos_leves', 'sum'),
        total_veiculos_envolvidos=('tipo_veiculo', 'count')
    )
    .reset_index()
)
acidentes_analise['total_feridos'] = (
    acidentes_analise['total_feridos_graves'] +
    acidentes_analise['total_feridos_leves']
)

print("\nInformações Principais dos Acidentes Mais Graves:")

acidente_mais_mortos_id = acidentes_analise[acidentes_analise['total_mortos'] == acidentes_analise['total_mortos'].max()]['id'].iloc[0]

for acc_id in [int(acidente_mais_mortos_id)]:
    agg_info = acidentes_analise[acidentes_analise['id'] == acc_id].iloc[0]
    print(f"Número de mortos: {int(agg_info['total_mortos'])}")
    print(f"Número de feridos (graves + leves): {int(agg_info['total_feridos'])}")
    print(f"Número de envolvidos (registros de pessoas/veículos): {int(agg_info['total_veiculos_envolvidos'])}")

    detail_info = nova_iguacu_df[nova_iguacu_df['id'] == acc_id].iloc[0]
    print(f"id: {detail_info['id']}")
    print(f"BR: {detail_info['br']}")
    print(f"KM: {detail_info['km']}")
    print(f"Dia da Semana: {detail_info['dia_semana']}")
    print(f"Horário: {detail_info['horario']}")
    print(f"Causa do Acidente: {detail_info['causa_acidente']}")

    vehicle_types = nova_iguacu_df[nova_iguacu_df['id'] == acc_id]['tipo_veiculo'].unique()
    print(f"Tipos de Veículos Envolvidos: {', '.join(vehicle_types)}")
```

### Análise Geral Final

A análise dos dados de acidentes nas rodovias federais de Nova Iguaçu em 2025 mostrou que foram registrados 1.081 acidentes ao longo do ano.

A maior parte das ocorrências aconteceu na **BR-116**, que concentrou aproximadamente **72%** dos acidentes com **778 registros**. Em seguida aparecem a BR-465 com cerca de 17% e a BR-493 com aproximadamente 10%. Alguns trechos da BR-116, especialmente entre os quilômetros 181 e 187, apresentaram maior concentração de acidentes, indicando possíveis pontos críticos de trânsito.

Ao longo do ano, **novembro foi o mês com maior número de acidentes**, representando cerca de **12,2% do total de 132 registros**. Já outubro teve aproximadamente 6,7% e janeiro cerca de 6,8% , sendo os **meses com menor registro.**

No total, os acidentes resultaram em **17 mortes e 486 pessoas feridas**. Considerando os registros, cerca de 89,8% dos acidentes tiveram vítimas, enquanto aproximadamente 10,1% ocorreram sem vítimas.

Entre os tipos de acidentes, os **mais comuns foram colisão traseira**, responsável por cerca de **27,8% **dos casos , colisão lateral no mesmo sentido com aproximadamente **18,8%** e queda de ocupante de veículo, representando cerca de **18,3%**.

Em relação aos veículos envolvidos, **motocicletas em registros 414 vezes e automóveis 380**, sendo os tipos mais frequentes.

A análise também mostrou que a maior parte das pessoas envolvidas nos acidentes era do sexo masculino aproximadamente** 72%**, enquanto **17,8%** **eram do sexo feminino**.

Quanto ao momento das ocorrências, cerca de 60,3% dos acidentes aconteceram durante o dia e 30,9% durante a noite. Apesar de menos frequentes, **os acidentes noturnos concentraram aproximadamente 64,7% das mortes registradas**.

De forma geral, os dados indicam que os acidentes em Nova Iguaçu estão concentrados em rodovias de grande movimento, principalmente na BR-116, envolvendo com maior frequência motociclistas e condutores do sexo masculino, e podem estar relacionados a fatores como alto fluxo de veículos, comportamento de risco e atenção no trânsito.
